# Probabilidade - Uma introdução ao acaso.

```python
# Importando as bibliotecas necessárias

import numpy as np
from scipy import stats
import numpy as np
import matplotlib.pyplot as plt
```

1. Uma máquina embala ração para cachorro em pacotes de 15 kg, mas cada pacote produzido acaba tendo um peso aleatório, com distribuição Normal, com média 15kg, mas com desvio padrão 0,1 kg. Obtenha a probabilidade de que um pacote selecionado aleatoriamente dessa produção tenha mais que 15,2 kg.

```python
# Definindo os parâmetros da distribuição normal
mu = 15.0  # Média da distribuição
desvio_padrao = 0.1  # Desvio padrão da distribuição

# Criando um conjunto de valores de x para o gráfico
x = np.linspace(mu - 3*desvio_padrao, mu + 3*desvio_padrao, 100)

# Calculando os valores de densidade de probabilidade para cada valor de x
y = stats.norm.pdf(x, mu, desvio_padrao)

# Criando uma figura e um eixo para o gráfico
fig, ax = plt.subplots(figsize=(10, 2))

# Plotando a curva da distribuição normal
ax.plot(x, y)

# Definindo o limite superior para o cálculo da probabilidade
limite_superior = 15.2

# Calculando a probabilidade de um valor estar abaixo do limite superior
probabilidade = 1 - stats.norm.cdf(limite_superior, loc=mu, scale=desvio_padrao)

# Configurando a largura do intervalo para sombrear a área
std_lim = 1  # 68% CI

# Calculando o limite inferior do intervalo
limite_inferior = mu - std_lim * desvio_padrao

# Preenchendo a área entre o limite inferior e o limite superior no gráfico
plt.fill_between(x, y, where=(limite_inferior < x) & (x < limite_superior), alpha=.5)

# Adicionando texto com os valores dos limites ao gráfico
plt.text(limite_inferior, 0.05, f'{limite_inferior:.2f}', ha='center')
plt.text(limite_superior, 0.05, f'{limite_superior:.2f}', ha='center')

# Imprimindo a probabilidade de que um pacote tenha mais de 15,2 kg
print(f'A probabilidade de um pacote ter mais que {limite_superior:.2f} kg é de {probabilidade * 100:.2f}%')

# Exibindo o gráfico
plt.show()
```

2. Da máquina descrita acima, obtemos o percentil de ordem 95% para o peso dos pacotes. Ou seja, o valor tal que a probabilidade de um pacote ter peso menor que este seja de 95%.

```python
# Definindo os parâmetros da distribuição normal
# Calculando o percentil de ordem 95%
media = 15.0  # Média da distribuição
desvio_padrao = 0.1  # Desvio padrão da distribuição
percentil_95 = stats.norm.ppf(0.95, loc=media, scale=desvio_padrao)

# Criando um conjunto de valores de x para o gráfico
x = np.linspace(media - 3*desvio_padrao, media + 3*desvio_padrao, 100)

# Calculando os valores de densidade de probabilidade para cada valor de x
y = stats.norm.pdf(x, media, desvio_padrao)

# Criando uma figura e um eixo para o gráfico
fig, ax = plt.subplots(figsize=(10, 2))

# Plotando a curva da distribuição normal
ax.plot(x, y)

# Adicionando uma linha vertical no percentil de ordem 95%
ax.axvline(x=percentil_95, color='r', linestyle='--', label=f'Percentil 95%: {percentil_95:.2f} kg')

# Adicionando uma legenda ao gráfico
ax.legend()

# Exibindo o gráfico
plt.show()
```

3. Supomos duas variáveis aleatórias $Y_1$ e $Y_2$, a primeira com distribuição Normal Padrão (média=0, variância=1), a segunda com distribuição *t-studdent* com 5 graus de liberdade (também padrão). Calculamos, para cada uma, a probabilidade de ser menor que cada um dos valores contidos no objeto $Y$ na célula abaixo:

```python
# Definindo o objeto Y
Y = np.linspace(-6, 6, 200)

# Criando as distribuições para Y1 (Normal Padrão) e Y2 (t-Student com 5 graus de liberdade)
Y1 = stats.norm(0, 1)
Y2 = stats.t(5)

# Calculando as probabilidades para cada valor em Y para Y1 e Y2
probabilidade_Y1 = [Y1.cdf(y) for y in Y]
probabilidade_Y2 = [Y2.cdf(y) for y in Y]
```

4. Uma visualização gráfica dos dados do item anterior.

```python
# Criando uma figura e um eixo para o gráfico
fig, ax = plt.subplots(figsize=(10, 5))

# Plotando as probabilidades de Y1 e Y2 em relação a Y
ax.plot(Y, probabilidade_Y1, label='$Y_1$ (Normal Padrão)')
ax.plot(Y, probabilidade_Y2, label='$Y_2$ (t-Student)')

# Adicionando rótulos aos eixos
ax.set_xlabel('Valores de Y')
ax.set_ylabel('Probabilidade')

# Adicionando uma legenda ao gráfico
ax.legend()

# Exibindo o gráfico
plt.title('Probabilidade de $Y_1$ e $Y_2$ serem menores que valores em Y')
plt.grid(True)
plt.show()
```

5. 3. Supondo duas variáveis aleatórias $Y_1$ e $Y_2$, a primeira com distribuição Normal Padrão (média=0, variância=1), a segunda com distribuição *t-studdent* com 5 graus de liberdade (também padrão). Calculamos, para cada uma, o percentil (a inversa da *CDF*) para cada um dos valores do objeto *P* abaixo:

```python
Y=np.linspace(0, 1, 20)
# Seu código começa aqui
import numpy as np
from scipy import stats

# Defina o objeto P
P = np.linspace(0, 1, 20)

# Crie as distribuições para Y1 (Normal Padrão) e Y2 (t-Student com 5 graus de liberdade)
Y1 = stats.norm(0, 1)
Y2 = stats.t(5)

# Calcule os percentis para cada valor em P para Y1 e Y2
percentis_Y1 = [Y1.ppf(p) for p in P]
percentis_Y2 = [Y2.ppf(p) for p in P]
```

6. Um gráfico com os dados do item anterior.

```python
# Criando uma figura e um eixo para o gráfico
fig, ax = plt.subplots(figsize=(10, 5))

# Plotando os percentis de Y1 e Y2 em relação a P
ax.plot(P, percentis_Y1, label='$Y_1$ (Normal Padrão)')
ax.plot(P, percentis_Y2, label='$Y_2$ (t-Student)')

# Adicionando rótulos aos eixos
ax.set_xlabel('Valores de P')
ax.set_ylabel('Percentil')

# Adicionando uma legenda ao gráfico
ax.legend()

# Exibindo o gráfico
plt.title('Percentis de $Y_1$ e $Y_2$ em relação a P')
plt.grid(True)
plt.show()
```
